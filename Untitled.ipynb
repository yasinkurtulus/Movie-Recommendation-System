```python
import pandas as pd
import numpy as np
import os 
```

```python
movies=pd.read_csv('data/tmdb_5000_movies.csv')
credits=pd.read_csv('data/tmdb_5000_credits.csv')
```

```python
movies.head(2)
```

```python
credits.head(2)
```

```python
movies=pd.merge(movies, credits, left_on='id', right_on='movie_id')
```

```python
movies.shape
```

```python
movies=movies.drop("title_x",axis=1)
```

```python
movies["original_language"].value_counts()
```

```python
movies[movies["original_language"]=="tr"]
```

```python
movies=movies[["id","original_title","overview","genres","keywords","cast","crew"]] 
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
#drop null values
movies.dropna(inplace=True)
```

```python
movies.iloc[0]["genres"]
#it is a string
```

```python
import ast
# ast literal_eval help us to convert sting into list
def convert(text):
    l=[]
    for i in ast.literal_eval(text):
        l.append(i["name"])
    return l
    
```

```python
movies["genres"]=movies["genres"].apply(convert)
```

```python
movies.head()
```

```python
movies.iloc[0]["keywords"]
#it is a string
```

```python
movies["keywords"]=movies["keywords"].apply(convert)
```

```python
movies.head()
```

```python
# we will take first 3 cast
def convert_cast(text):
    l=[]
    counter=0
    for i in ast.literal_eval(text):
        if counter <3:
            l.append(i["name"])
        counter+=1
    return l
    
```

```python
movies["cast"]=movies["cast"].apply(convert_cast)
```

```python
movies.iloc[0]["cast"]
```

```python
def convert_director(text):
    l=[]
    for i in ast.literal_eval(text):
        if i["job"]=="Director":
            l.append(i["name"])
            break
    return l
```

```python
movies["crew"]=movies["crew"].apply(convert_director)
```

```python
movies["overview"]=movies["overview"].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#we will drop the space betwween words
#Sam Worthington -> SamWorthington

def remove_space(text):
    l=[]
    for i in text:
        l.append(i.replace(" ",""))
    return l
```

```python
movies["crew"]=movies["crew"].apply(remove_space)
movies["cast"]=movies["cast"].apply(remove_space)
movies["genres"]=movies["genres"].apply(remove_space)
```

```python
movies["keywords"]=movies["keywords"].apply(remove_space)
```

```python
movies.head()
```

```python
movies["tags"]=movies["overview"]+movies["genres"]+movies["keywords"]+movies["cast"]+movies["crew"]
```

```python
movies.head()
```

```python
new_df=movies[["id","original_title","tags"]]
```

```python
new_df.head()
```

```python
new_df["tags"]=new_df["tags"].apply(lambda x: " ".join(x))
```

```python
new_df.iloc[0]["tags"]
```

```python
new_df["tags"]=new_df["tags"].apply(lambda x:x.lower())
```

```python
new_df.iloc[0]["tags"]
```

```python
#we will use stemmer befor count vectorization,
#for example there are love,loved loving, all of them have the same meaning so we need to convert them into love
```

```python
!pip install nltk
```

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
#this function find the root of each word
def stems(text):
    l=[]
    for i in text.split():
        l.append(ps.stem(i))
    return " ".join(l)
```

```python
new_df["tags"]=new_df["tags"].apply(stems)
```

```python
new_df.iloc[0]["tags"]
```

```python
!pip install scikit-learn
```

```python
import sklearn
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
vector=cv.fit_transform(new_df["tags"]).toarray()
```

```python
vector
```

```python
#it means ve have 4800 document and we choose top 5000 most frequent
vector.shape
```

```python
cv.get_feature_names_out()
```

```python
#and now we will use cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
similarity.shape
```

```python
new_df[new_df["original_title"]=="Spider-Man"].index[0]
```

```python
def recommend(movie):
    index=new_df[new_df["original_title"]==movie].index[0]
    distances=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].original_title)
```

```python
recommend("Spider-Man")
```
